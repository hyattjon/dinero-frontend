```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
font_path = '/Library/Fonts/latinmodern-math.otf' 
fm.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Latin Modern Math'
```

```python
data = pd.read_csv('./new_data/classified_transactions.csv')
data['date'] = pd.to_datetime(data['date'])
```

```python
data['spending'] = data['amount'] < 0
```

```python
val_dfs = {}
for val in data['category'].unique():
    print(val)
    val_dfs[val] = data[data['category'] == val]
    display(val_dfs[val].head())
```

```python
spending_df = data[data['spending'] == True]
```

```python
today = spending_df[spending_df['date'] >= '2025-01-01']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(today['date'], today['amount'])
plt.show()
```
